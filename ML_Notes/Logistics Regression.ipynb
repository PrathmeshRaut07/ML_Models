```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
#u can use label encoding to convert into 0 ,1...2,3
x=df.iloc[:,2:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
print(x_train)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
x_test[0]
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict([[-0.80480212,  0.50496393]])
y_pred
```

```python
y_pred=classifier.predict(sc_x.transform([[30,  87000]]))
y_pred
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
y_test
```

```python
dff=pd.DataFrame({'Real_value':y_test,'Predicted_value':y_pred})
dff.head(20)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
a=accuracy_score(y_test,y_pred)
a
```

```python
#Visualaizing the test result 
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = x_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

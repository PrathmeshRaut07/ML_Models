```python
import pandas as pd
import numpy as np
```

```python
#to predict salaries
df=pd.read_csv('Position_Salaries.csv')
df
```

# we came to know that he was a Region Mnaager 
we can consider his level 6.5

```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

# Creating Linear Regression model on the whole data set

```python
from sklearn.linear_model import LinearRegression
lin_req= LinearRegression()
lin_req.fit(x,y)
```

# Creating Polynomial Regression model on whole data set y=b0+b1x1+b2x1^2+.....bnx1^n-1

```python
from sklearn.preprocessing import PolynomialFeatures
poly_req=PolynomialFeatures(degree=2)#t_ill x^2
x_poly=poly_req.fit_transform(x)#x contains level that we need to transform
lin_req_2=LinearRegression()
lin_req_2.fit(x_poly,y)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

# Visulization of Linear Regression

```python
plt.scatter(x,y)
plt.plot(x,lin_req.predict(x),color='red')
plt.title('Truth or bLUFF/Linear Regressiom')
plt.xlabel('position')
plt.ylabel('salary')
```

```python
plt.scatter(x,y)
plt.plot(x,lin_req_2.predict(x_poly),color='red')
plt.title('Truth or bLUFF/Linear Regressiom')
plt.xlabel('position')
plt.ylabel('salary')
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_req=PolynomialFeatures(degree=4)#t_ill x^2
x_poly=poly_req.fit_transform(x)#x contains level that we need to transform
lin_req_2=LinearRegression()
lin_req_2.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,lin_req_2.predict(x_poly),color='red')
plt.title('Truth or bLUFF/Linear Regressiom')
plt.xlabel('position')
plt.ylabel('salary')
```

```python
#degree=4 curve is better 
```

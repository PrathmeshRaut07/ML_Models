```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Position_Salaries.csv')
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
plt.scatter(x,y)
plt.plot(x,regressor.predict(x),color='red')
plt.title('Truth or bLUFF/Tree Regressiom')
plt.xlabel('position')
plt.ylabel('salary')
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor   
regressor=RandomForestRegressor(n_estimators=10,random_state=0)#we need to endter parameter of number of trees
regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
plt.scatter(x,y)
plt.plot(x,regressor.predict(x),color='red')
plt.title('Truth or bLUFF/Forest Regressiom')
plt.xlabel('position')
plt.ylabel('salary')
```
